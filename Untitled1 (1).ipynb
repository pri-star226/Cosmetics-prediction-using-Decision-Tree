```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
import pickle as pkl
from sklearn import tree
```

```python
df = pd.read_csv("cosmetics.csv")
df.head()
```

```python
inputs = df.drop('Price',axis='columns')
```

```python
target = df['Price'] 
```

```python
le_Label= LabelEncoder()
le_Brand= LabelEncoder()
le_Name= LabelEncoder()
le_Ingredients= LabelEncoder()
le_Combination= LabelEncoder()
le_Rank= LabelEncoder()
le_Dry= LabelEncoder()
le_Normal= LabelEncoder()
le_Oily= LabelEncoder()
le_Sensitive= LabelEncoder()
```

```python
inputs['Label_n'] = le_Label.fit_transform(inputs['Label'])
inputs['Brand_n'] = le_Brand.fit_transform(inputs['Brand'])
inputs['Name_n'] = le_Name.fit_transform(inputs['Name'])
inputs['Rank_n'] = le_Rank.fit_transform(inputs['Rank'])
inputs['Ingredients_n'] = le_Ingredients.fit_transform(inputs['Ingredients'])
inputs['Combination_n'] = le_Combination.fit_transform(inputs['Combination'])
inputs['Dry_n'] = le_Dry.fit_transform(inputs['Dry'])
inputs['Normal_n'] = le_Normal.fit_transform(inputs['Normal'])
inputs['Oily_n'] = le_Oily.fit_transform(inputs['Oily'])
inputs['Sensitive_n'] = le_Sensitive.fit_transform(inputs['Sensitive'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['Label','Brand','Name','Ingredients','Combination','Rank','Dry','Normal','Oily','Sensitive'],axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs_n, target, test_size=0.2, random_state=42)
```

```python
model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
```
